```python
# importing skincare_products_clean.csv

skincare = pd.read_csv("C:\\Users\\HP\\OneDrive\\Desktop\\skincare-analysis\\data\\raw\\Skincare_cleandataset\\skincare_products_clean.csv")
skincare.head()
```

```python
# Filtering Sephora Dataset to only Skincare

## Finding the skincare category
sephora['primary_category'].value_counts().head(10)
```

```python
## Creating a new dataframe that only contains skincare items

skincare_sephora = sephora[
    sephora['primary_category']
    .str.strip()
    .str.lower()
    .str.contains('skincare')
].copy()

skincare_sephora.shape
```

```python
##Finding type of skincare

skincare_sephora['secondary_category'].value_counts().head(10)
```

```python
# Saving only skincare data

skincare_sephora.to_csv("C:\\Users\\HP\\OneDrive\\Desktop\\skincare-analysis\\data\\processed\\skincare_sephora.csv")
```

```python
# Brand Performance
## Which brands are most loved?

brand_perf = (
    skincare_sephora
    .groupby('brand_name')
    .agg(
        num_products=('product_id', 'count'),
        avg_price=('price_usd', 'mean'),
        avg_rating=('rating', 'mean'),
        total_loves=('loves_count', 'sum')
    )
    .sort_values('total_loves', ascending=False)
)

brand_perf.head(10)
```

```python
# Saving the table

brand_perf.to_csv("C:\\Users\\HP\\OneDrive\\Desktop\\skincare-analysis\\data\\processed\\brand_performance.csv")
```

```python
#Product Performance
## What products are most loved?

top_products = (
    skincare_sephora
    .sort_values('loves_count', ascending=False)
    [['product_name', 'brand_name', 'price_usd', 'rating', 'loves_count']]
    .head(10)
)

top_products.head(10)
```

```python
# Saving the table

top_products.to_csv("C:\\Users\\HP\\OneDrive\\Desktop\\skincare-analysis\\data\\processed\\top_products.csv",index=False)
```

```python
#From the skincare dataframe

##Cleaning the price column

import pandas as pd

skincare['price'] = (
    skincare['price']
    .str.replace('$', '', regex=False)
    .str.replace('£', '', regex=False)
    .str.strip()
)

skincare['price'] = pd.to_numeric(skincare['price'], errors='coerce')
```

```python
skincare['price'].describe()
```

```python
# Finding the most common ingredients used

from collections import Counter

ingredients_series = (
    skincare['clean_ingreds']
    .str.lower()
    .str.split(',')
)

ingredient_counts = Counter(
    i.strip() for sublist in ingredients_series for i in sublist
)

ingredient_counts.most_common(15)
```

```python
# Saving the result

## Converting to a dataframe

import pandas as pd

ingredient_df = pd.DataFrame(
    ingredient_counts.items(),
    columns=["ingredient", "count"]
)

## Saving as a csv

ingredient_df.to_csv("C:\\Users\\HP\\OneDrive\\Desktop\\skincare-analysis\\data\\processed\\ingredient_usage.csv" , index = False)
```
